```python
import pandas as pd
```

```python
data = {
    'Name':['Alice' , 'Bob' , 'Charline' , 'David'],
    'Age':[25,30,17,28],
    'City':['New york' , 'San francisco' , 'Alexandira' , 'New york'],
    'Salary':[100,150,200,300]
}
```

```python
df = pd.DataFrame(data)

df
```

```python
print('The minimum salary is' , df[['Salary']].min())
print('The maximum salary is' , df[['Salary']].max())
print('The mean age is' , df[['Age']].mean())
```

```python
ndata = [10,20,30,40,50]
s = pd.Series(ndata)

print(s.min())
print(s.max())
print(s.mean())
```
